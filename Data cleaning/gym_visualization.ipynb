```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=  pd.read_csv('gym_analysis.csv')
```

```python
# Plot a histogram for age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=15, kde=True, color='blue')
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate average weight for each gender
gender_avg_weight = df.groupby('Gender')['Weight'].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Weight', data=gender_avg_weight, palette='viridis')
plt.title('Average Weight by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Weight (kg)', fontsize=12)
plt.show()
```

```python
# Scatter plot for Calories Burned vs. Session Duration
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Session_Duration', y='Calories_Burned', hue='Gender', data=df, alpha=0.7)
plt.title('Calories Burned vs. Session Duration', fontsize=16)
plt.xlabel('Session Duration (hours)', fontsize=12)
plt.ylabel('Calories Burned', fontsize=12)
plt.legend(title='Gender')
plt.grid(alpha=0.7)
plt.show()
```

```python
# Bar plot for Workout Frequency grouped by Experience Level
plt.figure(figsize=(8, 6))
sns.boxplot(x='Experience_Level', y='Workout_Frequency', data=df, palette='cool')
plt.title('Workout Frequency by Experience Level', fontsize=16)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Workout Frequency (days/week)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count plot of the most common workout types
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='Workout_Type', order=df['Workout_Type'].value_counts().index, palette='Set2')
plt.title('Most Common Workout Types', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Workout Type', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar chart for average calories burned by experience level
avg_calories_by_exp = df.groupby('Experience_Level')['Calories_Burned'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Experience_Level', y='Calories_Burned', data=avg_calories_by_exp, palette='viridis')
plt.title('Average Calories Burned by Experience Level', fontsize=16)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Average Calories Burned', fontsize=12)
plt.show()
```
